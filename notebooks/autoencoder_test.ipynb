```python
def pad_sequences(data, maxlen, padding_value=0):
    padded_data = np.full((len(data), maxlen, data[0].shape[0]), padding_value)
    # print(padded_data.shape, len(data), data[0].shape)
    for i, seq in enumerate(data):
        length = seq.shape[1]
        # print(padded_data[i, :length, :].shape, seq.shape)
        padded_data[i, :length, :] = seq.T
    return padded_data

def process_lengths(predictors):

    apoyos_sizes = pd.Series([apo.shape[1] for apo in predictors])

    q5 = apoyos_sizes.quantile(0.05)
    q95 = apoyos_sizes.quantile(0.95)

    idx = (apoyos_sizes > q5)*(apoyos_sizes < q95)

    filtered_apoyos = []

    maxlen = 0

    for apo in predictors:
        
        cond = (apo.shape[1] > q5)*(apo.shape[1] < q95)

        if cond:
            filtered_apoyos.append(apo)
            if apo.shape[1] > maxlen:
                maxlen = apo.shape[1]
            
    # print(len(predictors_apoyos), len(filtered_apoyos), maxlen)

    # Pad sequences to a fixed length for batch processing
    padded_point_sets = pad_sequences(filtered_apoyos, maxlen)
    
    return padded_point_sets
```

```python
from tensorflow.keras.layers import Input, Dense, GlobalMaxPooling3D, AveragePooling3D, MaxPooling3D, GlobalMaxPooling1D

class ReshapeLayer(Layer):
    def __init__(self, target_shape):
        super(ReshapeLayer, self).__init__()
        self.target_shape = target_shape

    def call(self, inputs):
        return tf.reshape(inputs, self.target_shape)

def build_pointnet(input_shape):
    
    input_layer = Input(shape=input_shape)
    masked_layer = Masking(mask_value=0)(input_layer)
    
    # PointNet layers
    # x = AveragePooling3D(pool_size=5)(masked_layer)
    x = Dense(64, activation='relu')(masked_layer)
    x = Dense(128, activation='relu')(x)
    x = Dense(1024, activation='relu')(x)
    x = GlobalMaxPooling1D()(x)
    # x = GlobalMaxPooling3D()(x)
    x = Dense(512, activation='relu')(x)
    x = Dense(256, activation='relu')(x)
    encoded = Dense(128, activation='relu')(x)
    
    x = Dense(256, activation='relu')(encoded)
    x = Dense(512, activation='relu')(x)
    x = Dense(1024, activation='relu')(x)
    x = Dense(np.prod(input_shape), activation='sigmoid')(x)
    
    decoded = ReshapeLayer((-1, input_shape[0], input_shape[1]))(x)
    
    autoencoder = Model(input_layer, decoded)
    
    learning_rate = 0.001  # Specify your learning rate here
    momentum = 0.6  # Specify your momentum here

    optimizer = tf.keras.optimizers.RMSprop(learning_rate=learning_rate, momentum=momentum)

    autoencoder.compile(optimizer=optimizer, loss='mse')
    
    return autoencoder
```
